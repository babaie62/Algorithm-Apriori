```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd 
import numpy as np
```

```python
dataset=pd.read_csv('dataset.csv')

transactions_ar = dataset.as_matrix()
transactions =transactions_ar.tolist()
for i in range (len(transactions)):
    for j in range(len(transactions[i])):
        if type(transactions[i][j])== float:
                   transactions[i][j]=''
                
transactions[1][0]
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
te_ary
```

```python
df1 = pd.DataFrame(te_ary, columns=te.columns_)
df=df1.drop(columns=[''])
df
```

```python
# apriori
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
print(frequent_itemsets) #dataframe with the itemsets
```

```python
#association_rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
print(rules)
```

```python
lift = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(lift) #dataframe with confidence, lift, conviction and leverage metrics calculated
```
